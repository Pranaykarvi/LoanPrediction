```python
house_num={
    'RENT':0,
    'OWN':1,
    'MORTGAGE':2,
    'OTHER':3
}
loanData_test['person_home_ownership']=loanData_test['person_home_ownership'].replace(house_num)
cb_person_default_on_file_num={
   'Y':0,
    'N':1,


}
loanData_test['cb_person_default_on_file']=loanData_test['cb_person_default_on_file'].replace(cb_person_default_on_file_num)
loan_grade_num={
    'A':0,
    'B':1,
    'C':2,
    'D':3,
    'E':4,
    'F':5,
    'G':6,


}
loanData_test['loan_grade']=loanData_test['loan_grade'].replace(loan_grade_num)
loan_intent_num={
    'DEBTCONSOLIDATION':0,
    'PERSONAL':1,
    'EDUCATION':2,
    'MEDICAL':3,
    'VENTURE':4,
    'HOMEIMPROVEMENT':5,

}
loanData_test['loan_intent']=loanData_test['loan_intent'].replace(loan_intent_num)

loanData_test.head()
```

```python
loanData_test.shape
```

```python

import pandas as pd
import lightgbm as lgb
from catboost import CatBoostClassifier
from sklearn.ensemble import VotingClassifier



# For the test set, extract features (excluding 'id')
X_test = loanData_test.copy()

# LightGBM model
params = {
    'objective': 'binary',
    'boosting_type': 'gbdt',
    'metric': 'binary_error',  # Binary classification
    'num_leaves': 31,          # Controls the complexity of the model
    'learning_rate': 0.05,     # Lower learning rate for better accuracy
    'feature_fraction': 0.8     # Use a subset of features for each iteration
}
lgb_model = lgb.LGBMClassifier(**lgb_params)

# CatBoost model
catboost_params = {
    'iterations': 1000,
    'learning_rate': 0.05,
    'depth': 6,
    'loss_function': 'Logloss',
    'eval_metric': 'Accuracy',
    'early_stopping_rounds': 100
}
catboost_model = CatBoostClassifier(**catboost_params, verbose=0)

# Create the ensemble model using VotingClassifier (soft voting for probabilities)
ensemble_model = VotingClassifier(
    estimators=[('lgb', lgb_model), ('catboost', catboost_model)],
    voting='soft'  # Soft voting to average probabilities
)

# Train the ensemble model on the full training data
ensemble_model.fit(X, Y)

# Predict on the test set using the ensemble model
y_pred = ensemble_model.predict(X_test)

# Create a DataFrame with the test 'id' and predicted 'loan_status'
submission_df = pd.DataFrame({
    'id':loanData_test['id'],
    'loan_status': y_pred
})

# Save predictions to a CSV file
submission_df.to_csv('predicted_loan_status_Using_Ensemble.csv', index=False)

print("Ensemble prediction complete and saved to 'predicted_loan_status_Using_Ensemble.csv'")







```
